```python
# Importing Libraries
import codecademylib3_seaborn
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from svm_visualization import draw_boundary
from players import aaron_judge, jose_altuve, david_ortiz


fig, ax = plt.subplots()

# Viewing data on Aaron Judge
print(aaron_judge)

# Getting columns of Aaron Judge dataframe
print(aaron_judge.columns)

# Viewing unique enteries in description column
print(aaron_judge.description.unique())

# Viewing unique enteries in type column
print(aaron_judge.type.unique())

# Mapping strike and ball values in type to 1 and 0 respectively
aaron_judge['type'] = aaron_judge['type'].map({'S' : 1, 'B' : 0})
print(aaron_judge.type.unique())

# Viewing plate_x column of Aaron Judge Dataframe
print(aaron_judge.plate_x)

# Dropping null enteries in plate_x, plate_z and type columns
aaron_judge = aaron_judge.dropna(subset = ['plate_x', 'type', 'plate_z'])

# Scatterplot of plate_x and plate_y columns
plt.scatter(aaron_judge.plate_x, aaron_judge.plate_z, c = aaron_judge.type, cmap = plt.cm.coolwarm, alpha = 0.25)
plt.xlabel('Horizontal dispalcement of Pitch')
plt.ylabel('Vertical dispalcement of Pitch')
plt.title('Location of Pitch')

# Getting Training and Validation set
training_set, validation_set = train_test_split(aaron_judge, random_state = 1)

# Initialize an SVC object as classifier
classifier = SVC(kernel = 'rbf', gamma = 8, C=5)

# Train classifier on training set
classifier.fit(training_set[['plate_x', 'plate_z']], training_set[['type']])

# Setting size of axes
ax.set_ylim(-2, 4)

# Calling draw_boundary function
draw_boundary(ax, classifier)
plt.show()

# Validation score
print(classifier.score(validation_set[['plate_x', 'plate_z']], validation_set[['type']]))

# Putting everything into a function
def svm_classification(df):
    fig, ax = plt.subplots()
    df['type'] = df['type'].map({'S' : 1, 'B' : 0})
    df = df.dropna(subset = ['plate_x', 'type', 'plate_z'])
    plt.scatter(df.plate_x, df.plate_z, c = df.type, cmap = plt.cm.coolwarm, alpha = 0.25)
    plt.xlabel('Horizontal dispalcement of Pitch')
    plt.ylabel('Vertical dispalcement of Pitch')
    plt.title('Location of Pitch')
    training_set, validation_set = train_test_split(df, random_state = 1)
    classifier = SVC(kernel = 'rbf', gamma = 8, C= 5)
    classifier.fit(training_set[['plate_x', 'plate_z']], training_set[['type']])
    ax.set_ylim(-2, 6)
    ax.set_xlim(-3, 3)
    draw_boundary(ax, classifier)
    plt.show()
    print(classifier.score(validation_set[['plate_x', 'plate_z']], validation_set[['type']]))

svm_classification(david_ortiz)
svm_classification(jose_altuve)
```
